```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_breast_cancer
ld=load_breast_cancer()
ld.target
```

```python
df=pd.DataFrame(ld.data,columns=ld.feature_names)
df['target']=ld.target
```

```python
pd.set_option('display.max_columns',None)
df.head()
```

```python
df['target'].value_counts()
```

```python
x=df.drop(['target'],axis=1)
y=df['target']
```

```python
x
```

```python
for i in df:
    sns.boxplot(df[i])
    plt.show()
```

```python
df.corr().round()
```

```python
import statsmodels.formula.api as smf
model=smf.ols(formula='y~x',data=df).fit()
model.summary()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=700)
```

```python
from sklearn.preprocessing import RobustScaler
rs=RobustScaler()
```

```python
rs.fit(x_train)
```

```python
x_train = rs.transform(x_train)
x_test = rs.transform(x_test)
```

```python
from sklearn.naive_bayes import GaussianNB
gb = GaussianNB()
gb.fit(x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
print(accuracy_score(y_test,gb.predict(x_test)))
print(accuracy_score(y_train,gb.predict(x_train)))
```

```python
confusion_matrix(y_test,gb.predict(x_test))
```

```python
confusion_matrix(y_train,gb.predict(x_train))
```

```python
from sklearn.svm import SVC
model = SVC(kernel='rbf')
model.fit(x_train,y_train)
```

```python
## svm by default kernel -= rbf
#linear,centroid,ward,rdf are kernels
```

```python
confusion_matrix(y_test,model.predict(x_test))
```

```python
confusion_matrix(y_train,model.predict(x_train))
```
